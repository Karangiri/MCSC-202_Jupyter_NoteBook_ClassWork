## Question No. 2: The distance travelled by a ball falling in the air is given by $x(t)=x_0 + v_0t+0.5gt^2$. Plot $x$ versus $t$ for $t\in[0, 3]$ when $x_0=10$ $m$, $v_0=15$ $m/s$ and $g=-9.81$ $m/s^2$. Also, write a text ball in the figure window. 

# Solution:

```python
import numpy as np
import matplotlib.pyplot as plt

x0 = 10
v0 = 15
g = -9.81
t = np.linspace(0,3,20)
x = x0 + v0*t + 0.5*g*t**2

# Plot
plt.plot(t,x,'r')
plt.title('Falling Ball')
plt.xlabel('Time')
plt.ylabel('Position')
plt.text(1.5,14,'Ball')
plt.show()
```

## Question No. 3: Define a curve $f:[0, 0.2]\rightarrow \mathbb{R}$ deifned by 
## $$f(t)=\sin (2\pi t + \cos (2\pi t))$$ 
## (a)  Plot the line graph of $f$ using $200$ points
## (b) Use the color red, marker \*, and line width 2.0
## (c) Use the $t-axis$ limit $0<=t<=0.2$ and $f-axis$ limit $-2.0<=f(t)<=2.0$
## (d) Use the x-label, y-label and title with fontsize $16.0$ and fontweight bold

# Solution:

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0,0.2,200)
f = np.sin(20*np.pi*t) + np.sin(60*np.pi*t)
plt.plot(t,f,'r', linewidth = 2.0)
plt.axis([0, 0.2, -2.0, 2.0])
plt.title('Plot of the curve',fontsize = 16,fontweight='bold')
plt.xlabel('x-values',fontsize = 16,fontweight='bold')
plt.ylabel('f-values',fontsize = 16,fontweight='bold')
plt.show()
    
```

## Question No. 4: Creat a plot of the function $f(x) = e^{-x/10}\sin\pi x$ and $g(x) = xe^{-x/3}$ over the interval $[0, 10]$. Include labels for the $x-axis$ and $y-axis$, and a legend explaining which line is which plot.

# Solution:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
f = np.exp(-x/10)
g = x*np.exp(-x/3)

# Plot
plt.plot(x,f,'r')
plt.plot(x,g,'b')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.legend(['f','g'])
```

## Question No. 7: The shape of a Limacon can be defined parametrically as 
## $$r = r_0+\cos\theta, x = r\cos\theta, y = r\sin\theta$$
## When $r_0=1$, this curve is called Cardioid. Use this definition to plot the shape of a Limacon for $r_0=0.8,1.0,1.2$. Be sure to use enough points that the curve is closed and appears smooth. Use legend in center of the graph to identify which curve is which.

# Solution:

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0,2*np.pi,100)
def Shape_Curve(r0):
    r = r0 + np.cos(theta)
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    return x,y
def legd(r0):
    if (0< r0 < 1 or r0 > 1):
        string = 'Limacon'
    elif (int(r0)==1):
        string = 'Cardioid'
    else:
        string = 'Unknown'
    return string
r0 = 0.8
x, y = Shape_Curve(r0)
plt.plot(x,y,'r')
plt.axis('equal')
plt.legend([legd(r0)])
plt.show()
```
